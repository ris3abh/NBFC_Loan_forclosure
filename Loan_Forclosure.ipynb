```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
df['LAST_RECEIPT_AMOUNT'] = imputer.fit_transform(df['LAST_RECEIPT_AMOUNT'].values.reshape(-1, 1))
df['DPD'] = df['DPD'].fillna(0.0)
df['SCHEMEID'] = imputer.transform(df['SCHEMEID'].values.reshape(-1, 1))
```

```python
l1=LabelEncoder()
#####################################################################
df['last_authorization_year']=((df['LAST_RECEIPT_DATE']-df['AUTHORIZATIONDATE'])/np.timedelta64(1,'Y')).fillna(0)
df['last_authorization_month']=((df['LAST_RECEIPT_DATE']-df['AUTHORIZATIONDATE'])/np.timedelta64(1,'M')).fillna(0)
df['last_authorization_date']=((df['LAST_RECEIPT_DATE']-df['AUTHORIZATIONDATE'])/np.timedelta64(1,'D')).fillna(0)
#####################################################################
df['last_interest_year']=((df['LAST_RECEIPT_DATE']-df['INTEREST_START_DATE'])/np.timedelta64(1,'Y')).fillna(0)
df['last_interset_month']=((df['LAST_RECEIPT_DATE']-df['INTEREST_START_DATE'])/np.timedelta64(1,'M')).fillna(0)
df['last_interest_date']=((df['LAST_RECEIPT_DATE']-df['INTEREST_START_DATE'])/np.timedelta64(1,'D')).fillna(0)
#####################################################################
df['auth_interest_year']=((df['AUTHORIZATIONDATE']-df['INTEREST_START_DATE'])/np.timedelta64(1,'Y')).fillna(0)
df['auth_interest_month']=((df['AUTHORIZATIONDATE']-df['INTEREST_START_DATE'])/np.timedelta64(1,'M')).fillna(0)
df['auth_interest_day']=((df['AUTHORIZATIONDATE']-df['INTEREST_START_DATE'])/np.timedelta64(1,'D')).fillna(0)
#####################################################################
df['TENOR_DIFF']=df['CURRENT_TENOR']-df['ORIGNAL_TENOR']
df['AMT_DIFF']=df['LOAN_AMT']-df['NET_DISBURSED_AMT']
df['CUSTOMERID'].fillna(12009999,inplace=True)
df['SCHEMEID'].fillna(10909999,inplace=True)
df['LAST_RECEIPT_AMOUNT'].fillna(df['LAST_RECEIPT_AMOUNT'].mean(),inplace=True)
df['TENURE_DIFF'] = df['BALANCE_TENURE'] - df['COMPLETED_TENURE']
df['PRE_EMI'] = df['PRE_EMI_DUEAMT'] == df['PRE_EMI_RECEIVED_AMT']
df['PRE_EMI_DIFF'] = df['PRE_EMI_DUEAMT'] - df['PRE_EMI_RECEIVED_AMT']
df['EMI'] = df['EMI_DUEAMT'] == df['EMI_RECEIVED_AMT']
df['EMI_DIFF'] = df['EMI_DUEAMT'] - df['EMI_RECEIVED_AMT']
df['PRINCIPAL_FIDD'] = df['OUTSTANDING_PRINCIPAL'] - df['PAID_PRINCIPAL']
```

```python
categories = ['PRE_EMI', 'EMI', 'SCHEMEID','AGREEMENTID']
numericals = ['LOAN_AMT', 'TENOR_DIFF', 'DUEDAY', 'NET_RECEIVABLE', 'OUTSTANDING_PRINCIPAL', 
         'PAID_PRINCIPAL', 'PAID_INTEREST', 'MONTHOPENING', 'TENURE_DIFF', 'DPD', 'FOIR', 'MOB', 
        'PRE_EMI_DIFF', 'EMI_DIFF', 'ORIGNAL_TENOR', 'CURRENT_TENOR', 'BALANCE_TENURE', 
          'PRE_EMI_RECEIVED_AMT', 'EMI_DUEAMT', 'EMI_RECEIVED_AMT','LAST_RECEIPT_AMOUNT', 'NET_LTV','COMPLETED_TENURE', 'PRE_EMI_DUEAMT']
```

```python
scaler = StandardScaler()
encoder = LabelEncoder()

for i in numericals:
    dff[i] = scaler.fit_transform(dff[i].values.reshape(-1, 1))
    
for i in categories:
    dff[i] = encoder.fit_transform(dff[i].astype(str).values.reshape(-1, 1))
```

```python
X = dff.drop(columns=['FORECLOSURE'])
y = dff['FORECLOSURE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
d_train = lgb.Dataset(X_train, label=y_train)

params = {}
params['learning_rate'] = 0.21
params['boosting_type'] = 'gbdt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 35
params['min_data'] = 50
params['max_depth'] = 8
model = lgb.train(params, d_train, 400)

y_pred = model.predict(X_test)

for i in range(len(y_pred)):
    if y_pred[i]>=.5:       
       y_pred[i]=1
    else:  
       y_pred[i]=0
print('Accuracy:', accuracy_score(y_pred, y_test.values))
print('AUC score:', roc_auc_score(y_test.values, y_pred))
```
